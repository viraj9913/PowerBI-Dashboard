## Job Adverts in UK from 2017 to 2022

```python
jobadverts_df=pd.read_excel('dataset/EmploymentInUKFrom-2017-2022.xlsx',sheet_name='Table 7',header=4)
jobadverts_df
```

```python
jobadverts_df['Summary Profession Category'].unique()
```

```python
jobadverts_df = jobadverts_df.melt(id_vars=["Country", "Summary Profession Category", "Detailed Profession Category"], 
                                         var_name="Month", value_name="JobAdverts")
# melting the month column to make it appear as row
# value_name creating the new column to add its respective value of melted column
jobadverts_df
```

__Final dataset prepared for Job Adverts accross country in UK from 2017 to 2022__



```python


jobadverts_df=jobadverts_df[['Month','Country','Summary Profession Category','Detailed Profession Category','JobAdverts']]

# saving the cleaned file into the directory
directory='CleanedData/'
filepath=os.path.join(directory,'JobMarketCountrywise.csv')
if not os.path.exists(directory):
    os.makedirs(directory)

jobadverts_df.to_csv(filepath,index=False)
jobadverts_df
```

## Dataset of the Job market after Covid-19
E.g. value like 120.6 means 20.6% more job adverts than feb 2020
 
 value like 188.2 means 188.2% more job adverts than feb 2020

 __OR__

 120.6 means 12,060 job adverts

 188.2 means 18,820 job adverts
 


```python
JobCategory_df=pd.read_excel('dataset/DifferentJobCategories.xlsx',sheet_name='Adverts by category Feb 2020',header=6)

JobCategory_df
```

__Final dataset of hte Job Market after Covid-19__

```python
# removing the records which holds the null values
JobCategory_df=JobCategory_df.dropna()
# dropping the unncessary columns
JobCategory_df=JobCategory_df.drop(['Notes','Unknown'],axis=1)

# scaling the column e.g. 120.8 as 12,080
cols=['Date']
# storing the column into another df to retrieve in the end
unscale_cols=JobCategory_df[cols]

# dropping the date column
JobCategory_df.drop(columns=unscale_cols,inplace=True)
# scaling the remaining columns
JobCategory_df=JobCategory_df*100

# insert date column back to first position
JobCategory_df.insert(0, 'Date', unscale_cols)

JobCategory_df
```

```python
# saving the file into the directory
directory='CleanedData/'
filepath=os.path.join(directory,'JobCategory.csv')
if not os.path.exists(directory):
    os.makedirs(directory)

JobCategory_df.to_csv(filepath,index=False)
JobCategory_df
```
